```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pdRec = pd.read_csv('Recipes_sort.csv', header = 0)
#pdDef = pd.read_csv('Defects_sort.csv', header = 0)
Rec=pdRec.to_numpy()
#HEATID,Material_Code,NAME,WEIGHT
#Def=pdDef.to_numpy()
#PIECE_ID,HEAT_ID,DEFECT_TYPE,DEFECT_NAME,DEFECT_GROUP_ID,DEFECT_GROUP_Name

#удаляем лишние строки(дубликаты) для одной и той же плавки
heat=list(Rec[:,0])
m_code=list(Rec[:,1])
weight=list(Rec[:,3])
l=len(heat)

tl=[]#код материала для текущей плавки
ti=[]#индекс для tl
tl.append(m_code[l-1])
ti.append(l-1)

for i in reversed(range(l-1)):
    if(heat[i]!=heat[i+1]):
        tl.clear()#сбрасываем при новой плавке
        ti.clear()
    else:    
        if (m_code[i] in tl):#если это одна и та же плавка, и код разных материалов вдруг совпадает
            print("Удаляем строку",i,"плавки",heat[i],"код материала",m_code[i])
            j=tl.index(m_code[i])#находим номер строки, которая уже записана была. i-я строка записана не будет
            j=ti[j]#достаем номер строки
            Rec[j,3]=weight[i]+weight[j]
            print("Удаляемая строка:",i,"совпала",j,"удалить",weight[i],"был вес",weight[j],"стал вес:",Rec[j,3])
            Rec=np.delete(Rec,(i),axis=0)
        else:     #код материалов не совпадает. Добавляем этот код тоже       
            tl.append(m_code[i])
            ti.append(i)

#Вывод максимального числа осей в одной плавке
heat=list(Rec[:,0])
m_code=list(Rec[:,1])
weight=list(Rec[:,3])
l=len(heat)

maxax=heat.count(heat[l-1])
imax=l-1
for i in reversed(range(l-1)):
    if(heat[i]!=heat[i+1]):    
        temp=heat.count(heat[i])
        if(temp>maxax):
            maxax=temp
            imax=i           
    heat.pop()
print("Максимальное число осей для одной плавки: ",maxax)

#Вывод материалов с maxax
heat=list(Rec[:,0])
maxheat=heat[imax]

found = maxheat in heat
while(found):
    p=heat.index(maxheat)
    print("Номер строки и код материала maxax",maxheat,p,Rec[p,1])
    heat[p]=0
    found = maxheat in heat  

#поиск общего количества осей во всех HEATID
m_code=list(Rec[:,1])
ta=[]#список кодов отличающихся материалов
l=len(m_code)

for i in reversed(range(l)):
    temp=m_code[i]
    m_code.pop()
    found= temp in m_code
    if (found==False):
        ta.append(temp)
print("Максимальное число осей: ",len(ta),"коды материалов: ",ta)
    
pdRec=pd.DataFrame(Rec,index=None,columns=['HEATID','Material_Code','NAME','WEIGHT'])
pdRec.to_csv(r'Recipes_del.csv',index=None,header=True)

pdRec.head(10)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pdRec = pd.read_csv('Recipes_del.csv', header = 0)
Rec=pdRec.to_numpy()

#вычисление процентного соотношения материалов Recipes_p
heat=list(Rec[:,0])
weight=list(Rec[:,3])
dens=[]
l=len(heat)

tl=[]#вес для текущей плавки
tl.append(weight[l-1])

for i in reversed(range(l-1)):
    if(heat[i]!=heat[i+1]):
        temp=np.sum(tl)
        for j in range(len(tl)):
            tl[j]=round(((tl[j]/temp)*100),2)
        dens.extend(tl)
        #print("Сумма для",heat[i+1],":",temp)
        tl.clear()#сбрасываем при новой плавке
    tl.append(weight[i])
temp=np.sum(tl)
for j in range(len(tl)):
    tl[j]=(tl[j]/temp)*100
dens.extend(tl)
dens.reverse()
dena=np.array(dens)
dena=np.reshape(dena,(len(dens),1))
#print(dena.shape)
#print(Rec.shape)

Rec=np.hstack((Rec,dena)) 

pdRec=pd.DataFrame(Rec,index=None,columns=['HEATID','Material_Code','NAME','WEIGHT','PROCENT'])
pdRec.to_csv(r'Recipes_p.csv',index=None,header=True)
print(Rec)
```
